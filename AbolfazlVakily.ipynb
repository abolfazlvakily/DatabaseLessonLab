<h2 dir='rtl'>جهت خراب نشدن تیبل اصلی یک تیبل موقت می سازیم</h2>

```python
backup_from_table = '''
select *
into #T
from FinalTransaction
'''
```

```python
cursor = conn.cursor()
cursor.execute(backup_from_table)
```

```python
update = '''
update #T
set 
    MainFrameFinal = cast(0 as varbinary)
where
    widthImg > 70 and 
    AvgSpeed <> 0 and
    len(FrameNames) > 20
'''
```

```python
cursor.execute(update)
```
